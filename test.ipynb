```python
np.median(data)
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
```

```python

# Calculate IQR
IQR = Q3 - Q1
IQR
```

```python
# Define whiskers (1.5 times IQR)
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR
```
